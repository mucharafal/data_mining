Notebook to extract field of study and change it into index. Create external file with mapping fos <-> index.

```python
import sys
import json
import time
from os import listdir
from os.path import isfile, join
import os
import pandas
import matplotlib.pyplot as plt
import numpy as np
import multiprocessing
import re
from numba import jit


def lines_generator():
    onlyfiles = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
    dblp_part_regex = re.compile("dblp_part*")
    part_files_filter_obj = filter(lambda file_name: dblp_part_regex.match(file_name), onlyfiles)
    part_files = list(part_files_filter_obj)
    for filename in part_files:
        with open(filename) as file_handle:
            for line in file_handle:
                yield line
```

```python
def get_map():
    fos_map = {}
    index = 0
    onlyfiles = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
    dblp_part_regex = re.compile("dblp_part*")
    part_files_filter_obj = filter(lambda file_name: dblp_part_regex.match(file_name), onlyfiles)
    part_files = list(part_files_filter_obj)
    for filename in part_files:
        print(filename)
        with open(filename) as file_handle:
            for line in file_handle:
                try:
                    paper = json.loads(line)
                    fos = paper["fos"]
                    for field in fos:
                        name = field["name"]
                        if name not in fos_map:
                            fos_map[name] = index
                            index += 1
                except KeyError:
                    pass
                except:
                    pass
    return fos_map
```

```python
fos_map = get_map()
```

```python
def map_fos(fos_map):
    onlyfiles = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
    dblp_part_regex = re.compile("dblp_part*")
    part_files_filter_obj = filter(lambda file_name: dblp_part_regex.match(file_name), onlyfiles)
    part_files = list(part_files_filter_obj)
    for filename in part_files:
        print(filename)
        with open(filename) as file_handle:
            with open("fos_indexed_" + filename, "w+") as dest_handle:
                for line in file_handle:
                    try:
                        paper = json.loads(line)
                        fos = paper["fos"]
                        for field in fos:
                            name = field["name"]
                            index = fos_map[name]
                            field["name"] = str(index)
                        dest_handle.write(json.dumps(paper) + "\n")
                    except KeyError:
                        pass
                    except:
                        pass
```
