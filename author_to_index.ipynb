Processing fos_indexed_... files to change authors fields into index and save authors data externally.

```python
import sys
import json
import time
from os import listdir
from os.path import isfile, join
import os
import pandas
import matplotlib.pyplot as plt
import numpy as np
import multiprocessing
import re
from numba import jit  
```

```python
def get_map():
    authors_id_tuple_map = {}
    onlyfiles = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
    dblp_part_regex = re.compile("fos_indexed_dblp_part*")
    part_files_filter_obj = filter(lambda file_name: dblp_part_regex.match(file_name), onlyfiles)
    part_files = list(part_files_filter_obj)
    for filename in part_files:
        print(filename)
        with open(filename) as file_handle:
            for line in file_handle:
                try:
                    paper = json.loads(line)
                    fos = paper["authors"]
                    for field in fos:
                        id = field["id"]
                        if id not in authors_id_tuple_map:
                            authors_id_tuple_map[id] = field
                except KeyError:
                    pass
                except:
                    pass
    return authors_id_tuple_map
```

```python
authors_id_name_map = get_map()
```

```python
index_to_id_map = {}
index = 0
for id in authors_id_name_map.keys():
    index_to_id_map[index] = id
    index += 1
    
```

```python
id_to_index_map = {value: key for key, value in index_to_id_map.items()}
```

```python
with open("authors_index_to_id.json", "w+") as des:
    des.write(json.dumps(index_to_id_map))
```

```python
def map_authors():
    onlyfiles = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
    dblp_part_regex = re.compile("fos_indexed_dblp_part*")
    part_files_filter_obj = filter(lambda file_name: dblp_part_regex.match(file_name), onlyfiles)
    part_files = list(part_files_filter_obj)
    for filename in part_files:
        print(filename)
        with open(filename) as file_handle:
            with open("author_indexed_" + filename, "w+") as dest_handle:
                for line in file_handle:
                    try:
                        paper = json.loads(line)
                        authors = paper["authors"]
                        new_authors = []
                        for author in authors:
                            id = author["id"]
                            index = id_to_index_map[id]
                            new_authors.append({"index": index})
                        paper["authors"] = new_authors
                        dest_handle.write(json.dumps(paper) + "\n")
                    except KeyError:
                        pass
                    except:
                        pass
```
